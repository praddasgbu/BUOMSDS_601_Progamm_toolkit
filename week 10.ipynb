````python
df1 = pd.DataFrame({"a": [1,2,3], "b": [4,5,6]})
df1
````

````python
df2 = pd.DataFrame({"a": [9,8,7], "c": [98, 23, 45]}, index=[0,2,3])
df2
````

## NaN

````python
df1["a"] + df2["a"]
````

## Replacing `NaNs` with zero after the operation
Here the missing index values are not considered at all

````python
(df1["a"] + df2["a"]).fillna(0.0)
````

## Replacing missing index values with zeros before the operation
Here missing index values are considered zero hence we will see value for each index

````python
df1["a"].add(df2["a"], fill_value=0)
````

### Same process using `fillna`
This is a longer process

````python
(df1["a"] + df2["a"]).fillna(df1["a"]).fillna(df2["a"])
````

## Compute on dataframes
1. Above we saw operation on series
2. Let's look at dataframes

````python
df1.add(df2)
````

````python
df1.add(df2, fill_value=0)
````

````python
df1
````

````python
df2
````

**Why we still see `NaN`?**
1. `fill_value=0` only kicks in when one side is missing. If both sides are missing, it still gives `NaN`.
2. Label 3, column b:
   - df1 has no row with label 3 → missing
   - df2 has no column b → missing
   - Both sides missing → NaN
3. Label 1, column c:
   - df1 has no column c → missing
   - df2 has no row with label 1 → missing
   - Both sides missing → NaN

````python
df1.add(df2, fill_value=0).fillna(0)
````

## Problem 2

**Concept:** *reading CSV data, matplotlib, line graph*

**Task:**

Write a function `read_and_plot_column(filename, column_name)` that:
* Opens a CSV file (do **not** use pandas)
* Extracts the values from the specified column
* Casts all values to `float`
* Plots the values using `plt.plot()`
* Adds a title
* Displays the plot

### Main

#### Using python built in file open module

````python
def read_and_plot_column(filename, column_name):
    """
    Reads & extracts values of the column name in the function from csv file and plots a line graph 
    """
    # Open file
    with open(filename) as fp:
        reader = fp.readlines()
        
        # Define column names
        header = reader[0].strip("\n").split(",")
        
        # find index of column name
        col = header.index(column_name)

        # Strip newline and split each row by comma to get individual values, excluding header (reader[0])
        values = [row.strip("\n").split(",") for row in reader[1:]]
        
        # Use the column position (col) to extract only the target column's value from each row
        value = [v[col] for v in values]

        # cast values to float
        y = np.array(value, dtype=float)

        # Plot line graph
        plt.plot(y)
        
        #Add title
        plt.title(f"Line Graph for file {filename} and Column {column_name}")
        plt.show()
        
read_and_plot_column("02215_weather.csv", "tempmax")
````

#### Using Python CSV Reader

````python
def read_and_plot_column2(filename, column_name):
    """
    Reads & extracts values of the column name in the function from csv file and plots a line graph 
    """
    # Open file
    with open(filename) as fp:
        # Extract file as csv
        reader = csv.DictReader(fp)
        
        # List comprehension to collect the values of column and covert to float
        y = np.array([row[column_name] for row in reader], dtype=float)
        
        # Plot line graph
        plt.plot(y)
        
        #Add title
        plt.title(f"Line Graph for file {filename} and Column {column_name}")
        plt.show()
````

````python
def read_csv_to_dict(filename):
    """
    reads a CSV file and stores the data in a dictionary
    """
    data = np.genfromtxt(filename, delimiter=',', names=True)

    # Accumulator to return the final result as a dictionary
    acc = {}
    
    # Loop over names/columns
    for name in data.dtype.names:
        
        # Assign the NumPy array of values for each column
        acc[name] = data[name]
    return acc
````

**Task:**


Write a function `filter_rows_matching_pattern(df, column, pattern)` that:
* Returns rows where the specified column contains the given pattern
* Handles missing values
* Prints the number of matching rows

**Sample Output:**

If your selected column contained matching text values, the following could be your output:
```

### Main

````python
def filter_rows_matching_pattern(df, column, pattern):
    """
    Returns rows where the specified column contains the given pattern & handles missing values
    """
    
    # Boolean condition / filter to match pattern using `.str.contains()
    # Boolean condition / filter to handle missing values using `na=False`
    pattern_matching_filter = df[column].str.contains(pattern, na=False)
    
    # print the number of matching rows
    print(f"Found {len(df[pattern_matching_filter])} records of {column} containing {pattern}")
    
    # Return the filtered dataframe
    return df[pattern_matching_filter]
````

````python
def create_df_categorical(df_source, column_name):
    """
    Creates a new DataFrame with counts of each category and percentage of each category.
    """
    
    # Count values of all the categories
    counts = df_source[column_name].value_counts()
    
    # Compute percentage of all the categories
    pct = df_source[column_name].value_counts(normalize=True)
    return pd.DataFrame({"count": counts, "pct": pct})
````

## Problem 17

**Concept:** *Filtering records by category.*


**Task:**

Write a function `filter_weather(df, conditions, temp_min, temp_max)` that:
* Filters rows where:
  * conditions contain the given string
  * `tempmax` ≤ `temp_max`
  * `tempmin` ≥ `temp_min`
* Prints intermediate counts
* Returns the filtered DataFrame

**Sample Output:**

If your data matched the filtering conditions step by step, the following could be your output:
```
4 records with conditions Blue
3 records with conditions Blue and tempmax <= 50
2 records with conditions Blue and 20 <= temp <= 50
Found 2 records with conditions of Blue and temp between 20 and 50.
```

### Main

````python
def filter_weather(df, conditions, temp_min, temp_max):
    """
    Filters rows where: conditions contain the given string, tempmax ≤ temp_max and tempmin ≥ temp_min
    """

    # Filter conditions
    condition_filter = df[df["conditions"].str.contains(conditions, na=False)]

    # Filter temp_max and conditions
    temp_max_filter = condition_filter[condition_filter["tempmax"] <= temp_max]

    #Filter temp_min, temp_max and conditions
    temp_min_filter = temp_max_filter[temp_max_filter["tempmin"] >= temp_min]

    
    print(f'{len(condition_filter)} records with conditions {conditions}')
    print(f'{len(temp_max_filter)} records with conditions {conditions} and tempmax <= {temp_max}')
    print(f'{len(temp_min_filter)} records with conditions {conditions} and {temp_min} <= temp <= {temp_max}')
    print(f'found {len(temp_min_filter)} records with conditions of {conditions} and temp between {temp_min} and {temp_max}')

    return temp_min_filter
````

````python
# looking for perfect running days:
df_result = filter_weather(df_weather, "Partially cloudy", 4, 15)
print(f"The best running days were:")
print(df_result[['tempmin', 'tempmax','conditions',]])
print()

# looking for perfect beach days:
df_result = filter_weather(df_weather, "Clear", 5, 30)
print(f"The best beach days were:")
print(df_result[['tempmin', 'tempmax','conditions',]])
````

**Concept:** *Creating DataFrame, standardizing values.*


**Task:**

Write a function `create_standardized_values(df_source, columns, row_index=None)` that:
* Creates a new DataFrame
* Standardizes values relative to a given row
* Uses the first row if none is provided
* Returns the new DataFrame

**Sample Output:**

If the following was your dataframe:
```
      high_value  low_value
2020        100        50
2021        120        55
2022        140        60
2023        160        70
2024        180        75
```
If you call the function
`create_standardized_values(df_example, ["high_value", "low_value"], 2020)`,
you should get the following as the output:
```
      high_value  low_value
2020         1.0       1.00
2021         1.2       1.10
2022         1.4       1.20
2023         1.6       1.40
2024         1.8       1.50

```

### Main

````python
def create_standardized_values(df_source, columns, row_index=None):
    """
    Creates a new DataFrame. Standardizes values relative to a given row. Uses the first row if none is provided. Returns the new DataFrame.
    """
    
    # Use the first row if index row is none i.e. not provided 
    if row_index is None:
        row_index = df_source.index[0]
    
    return df_source[columns] / df_source[columns].loc[row_index]
````
